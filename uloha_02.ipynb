## Úloha 02: Vizualizace dat

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("fish_data.csv")  # spotify most streamed songs

# histogram hmotnosti ryb
plt.figure(figsize=(6, 4))
plt.hist(df["weight"], bins=20, color="lightseagreen", edgecolor="teal")
plt.ylabel("Počet jedinců (ryb)")
plt.xlabel("Váha [kg]")
plt.grid(True)
plt.gca().set_facecolor("aliceblue")
plt.show()
```

```python
# délka vs. hmotnost
plt.figure(figsize=(10,6))
plt.scatter(df["length"], df["weight"], color="teal", alpha=0.7)
plt.xlabel("délka [cm]")
plt.ylabel("délka [g]")
plt.gca().set_facecolor("aliceblue")
plt.grid(True)
plt.show()
```

```python
# boxplot poměr hmotnosti k délce podle druhu
plt.figure(figsize=(10, 6))
df.boxplot(column="w_l_ratio", by="species", grid=False)
plt.suptitle('Boxplot poměru hmotnosti k délce podle druhu')
plt.xlabel("Druhy ryb")
plt.ylabel("Poměr hmotnosti k délce")
plt.xticks(rotation=45)
plt.show()
```

```python
# bar plot průměrného poměru hmotnosti k délce podle druhu

avg_wl_ratio = df.groupby('species')['w_l_ratio'].mean().sort_values()
plt.figure(figsize=(6,4))
avg_wl_ratio.plot(kind='bar', color=("lightseagreen"))
plt.xlabel("Druhy ryb")
plt.ylabel("Průměrný poměr hmotnost-délka")
plt.xticks(rotation=45)
plt.gca().set_facecolor("aliceblue")
plt.show()
```

```python
# 

avg_length_by_species = df.groupby('species')['length'].mean()

plt.figure(figsize=(6,4))
plt.plot(avg_length_by_species.index, avg_length_by_species.values, marker="o", color="teal", linestyle="-")
plt.xlabel("Druhy ryb")
plt.ylabel("Průměrná delka [cm]")
plt.grid(True)
plt.xticks(rotation=45)
plt.gca().set_facecolor("aliceblue")
plt.show()
```
